### Brownian motion simulation

This is a loose translation of my KCL second year project (code was supplied in Fortran) into python. Brownian motion is a simple physical process to simulate due to the lack of modelled inter-particle interactions. The model, using stochasticity to represent the cumulative effect of many influencing factors which are taken to be uncorrelated, has been used outside of physical science, for example modelling stock market fluctuations.


The physical system considered consists of a number of 'heavy', spherical particles in a liquid, with no external forces such as gravity operating. In 1827 Robert Brown observed apparently random motion of particles in such a system while studying pollen grains suspended in water using a microscope. Einstein provided a theoretical understanding of this Brownian motion and diffusion in general in terms of the molecular – kinetic theory of heat: the suspended particles are understood to be buffeted by many collisions of the liquid molecules, due to their thermal motion, leading to “random walk” motion of the suspended 
particles. 

For a small number of particles, it is possible to track each particle and calculate the effect of each individual collision using Newton's laws. However in systems of many particles it is not practical to do so: this would necessitate both (i) increasing by many orders of magnitude the number of particles considered to include the fluid particles and (ii) significantly reducing the time step of the simulation to reflect the characteristic time of such collisions. This would render simulation impractical in terms of CPU time, and are unnecessary for modelling the  long term (on the order of milliseconds to seconds) behaviour of the large particles. 

A water molecule may experience of the order of 10^11 collisions per millisecond (3): in practice an effective model can average the effect of all the collisions occurring within a comparable time step dt much larger than the characteristic time of the individual collisions: Thus the effect at each time step dt (in this case 10^-4 s < dt < 10^-5 s) is essentially random due to the large number of collisions.

Thus the effect of the collisions can be modelled as a stochastic force, where the average force is zero (no directional preference), but with the characteristic magnitude set by the temperature of the system and other model parameters - the result is the Langevin Equation, which is integrated numerically in the simulation using Verlet integration. See  https://en.wikipedia.org/wiki/Langevin_equation#Brownian_motion_as_a_prototype, https://en.wikipedia.org/wiki/Verlet_integration). The python implementation is reproduced below:

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

Scipy.stats is used to calculate the simulation diffusion constant via linear regression of MSD vs time. 

I have defined a single class for the model, with the state of the system represented as an n_particles x 4 (x, y, vx, vy) numpy array. This enables efficient updating at each timestep, and easy analysis using numpy's built in functions (looping over particles can often be avoided). 

```python
class BrownianModel2D:

    def __init__(self, particle_mass, kT, gamma, npart, dt, timesteps):
        self.particle_mass = particle_mass
        self.kT = kT
        self.gamma = gamma
        self.npart = npart
        self.dt = dt
        self.timesteps = timesteps

        self.noisescale = (((2.0 * self.gamma * self.kT * self.dt) ** (0.5)) / self.particle_mass)
        self.timefactor = (self.gamma * self.dt) / self.particle_mass

    def initialise(self):
        '''Model state represented as a numpy array, with
        rows representing different particles
        columns = [x, y, vx, vy] - all initialised at zero'''

        self.state = np.zeros((self.npart, 4))

    def update_timestep(self):
        '''Velocity-Verlet integrator, vectorised for efficiency'''

        x = self.state[:, 0]  # vector: x positions for all particles
        y = self.state[:, 1]  # vector: y positions for all particles
        vx = self.state[:, 2]  # vector: x velocities for all particles
        vy = self.state[:, 3]  # vector: y velocities for all particles

        # v(t+dt) = v(t) - drag(v(t)) + noise
        new_vx = vx - (vx * (self.timefactor)) + (np.random.randn(self.npart) * self.noisescale)
        self.state[:, 2] = new_vx
        new_vy = vy - (vy * (self.timefactor)) + (np.random.randn(self.npart) * self.noisescale)
        self.state[:, 3] = new_vy

        # x(t+dt) = x(t) + v(t+dt)*dt
        # With no position dependent forces, V.V. algorithm is much simpler
        self.state[:, 0] = x + (new_vx * self.dt)  # update x positions
        self.state[:, 1] = y + (new_vy * self.dt)  # update y positions

    def update_MSD(self):
        '''Calculate mean squared displacement (averaged over particles) from state'''

        x = self.state[:, 0]  # vector: x positions for all particles
        y = self.state[:, 1]  # vector: y positions for all particles

        MSD = np.sum((x ** 2) + (y ** 2)) / self.npart

        return MSD

    def run_simulation(self, plot_trajectories=0):

        self.time = np.arange(self.timesteps) * self.dt
        self.MSD = np.zeros(self.timesteps)

        if plot_trajectories > 0:
            self.trajectories = np.zeros((self.timesteps, plot_trajectories*2))

        for t in range(self.timesteps):

            self.update_timestep()
            self.MSD[t] = self.update_MSD()

            if plot_trajectories > 0:
                for p in range(plot_trajectories):
                    self.trajectories[t, 2 * p] = self.state[p, 0]
                    self.trajectories[t, (2 * p) + 1] = self.state[p, 1]

            if t % (self.timesteps/10) == 0:
                print 'running...', 100 * t / self.timesteps, '%'

        print 'Simulation complete\n'

        if plot_trajectories > 0:
            plt.rcParams["figure.figsize"] = [9, 9]
            for i in range(plot_trajectories):
                plt.plot(self.trajectories[:, (2 * i)],
                         self.trajectories[:, (2 * i) + 1],
                         lw=0.3)

            plt.axis('equal')
            plt.title('Trajectory for Brownian particle(s)')
            plt.xlabel('Displacement (m)')
            plt.ylabel('Displacement (m)')
            plt.savefig('brownian_trajectory.png')
            plt.show()

        return self.MSD, self.time
```

It is worth definining two further variables: noisescale and timefactor are used to set the characteristic scale of the stochastic force according to the parameters of the model and simplify the appearance of the time evolution algorithm.

initialise() starts the system with all particles at the origin, with zero velocity. Since the particles do not interact this has no effect on the dynamics and enables easier calculation of displacements. 

update_timestep() evolves the system forward by one time step (again see https://en.wikipedia.org/wiki/Langevin_equation#Brownian_motion_as_a_prototype, https://en.wikipedia.org/wiki/Verlet_integration)

run_simulation() updates the system for the defined number of time steps, with the option of plotting any number of particle trajectories.

Other methods can easily be added either to calculate other quantities of interest such as average energy.

```python
# Parameters

kB = 1.381 * (10**-23)  # Botlzmann constant in J/K

T = 300                 # Temperature of system in K
kT = kB * T
M = 1.5*(10**-10)       # suspension particle mass in kg
gamma = 8*(10**-7)      # fluid drag factor in Ns/m

npart = 1000            # number of particles
dt = 0.00002            # simulation timestep in seconds
timesteps = 50000       # total number of time steps
```

A few parameters need to be defined for the system. 

These are configurable - but dt should not be changed dramatically. It should be at least an order of magnitude less than the characteristic relaxation time (under friction) of the brownian particles, but reducing it further can introduce numerical errors and does not improve simulation performance.

The program calculates the diffusion coefficient from the system via two methods (a theoretical value vs a derived value from the simulation data) and compares them. Using more particles and evolving over a longer time period should improve the accuracy but increase the running speed (each should do so linearly). 

```python
# MAIN PROGRAM

brownian = BrownianModel2D(M, kT, gamma, npart, dt, timesteps)

print '2D Brownian Motion simulation'
print '(%d timesteps = %f seconds)' % (timesteps, (dt*timesteps))
print 'Model Parameters:'
print 'kT =', kT, 'J \nParticle mass =', M, 'kg'
print 'Drag factor (gamma) =', gamma, 'Ns/m\n'

brownian.initialise()

MSD, time = brownian.run_simulation(plot_trajectories=3)

gradient, _, _, _, _ = stats.linregress(
    time, MSD)
Dexp = gradient / 4     # based on prediction <MSD> = 2 * dimensions * D * t
difference = abs(100 * (1 - (Dexp / (kT / gamma))))
Dpred = kT / gamma

print 'Predicted value of diffusion constant D = kT/gamma =', Dpred, 'm**2/s\n'
print 'Measured/simulation value of D =', Dexp, 'm**2/s'
print 'This is within', difference, '% of the predicted value'
```

As expected, the particles display a random walk in 2D, with dynamics as predicted.
